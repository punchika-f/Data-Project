```python
import requests
import pandas as pd
```

```python
# import game detail Top 100 in 2 weeks ago
url = "https://steamspy.com/api.php?request=top100in2weeks"
reponse = requests.get(url)
data = reponse.json()
```

```python
# convert to dataFrame
df = pd.DataFrame(data).T
```

```python
df
```

```python
# import genre from appid
def get_genre(appid):
    url_detail = f"https://steamspy.com/api.php?request=appdetails&appid={appid}"
    try:
        r = requests.get(url_detail)
        return r.json().get("genre", "")
    except:
        return ""

# match genre into dataFrame
df["genre"] = df["appid"].apply(get_genre)
```

```python
df
```

```python
# select only necessary column
df = df[[ "name", "genre","price","average_forever","average_2weeks","developer","publisher","positive","negative"]]
```

```python
df
```

```python
# Null value
print(df.isnull().sum())
```

```python
# change data type to numeric
df["price"] = pd.to_numeric(df["price"])
df["average_forever"] = pd.to_numeric(df["average_forever"])
df["average_2weeks"] = pd.to_numeric(df["average_2weeks"])
df
```

```python
# turn price to usd
df["price"] = df["price"] *100
df
```

```python
# save csv file
df.to_csv("steam_top100_with_genre.csv", index=False)
print("Saved as steam_top100_with_genre.csv")
```

```python
df['positive_review_ratio'] = df['positive'] / (df['positive'] + df['negative'])
df
```

# Analysis data

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## top 10 game that has the most playtime in last 2 weeks

```python
# set up chart
sns.set(style="whitegrid", font_scale=1)

# top 10 game that has the most playtime in last 2 weeks
top10 = df.sort_values(by="average_2weeks", ascending=False).head(10)

# chart detail
plt.figure(figsize=(10,5))
sns.barplot(data=top10, x="average_2weeks", y="name", palette="viridis")
plt.title("Top 10 Games by playtime (Last 2 Weeks)")
plt.xlabel("playtime (minutes)")
plt.ylabel("Game name")
plt.tight_layout()
plt.show()
```

## Top 10 the most popular genres

```python
# split genre : separate each genre into a single row.
genre_series = df['genre'].dropna().str.split(', ').explode()
genre_series
```

```python
# count genre and select top 10 genre
genre_counts = genre_series.value_counts().head(10)
genre_counts
```

```python
# create chart Top 10 genre
plt.figure(figsize=(10,5))
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette="mako")
plt.title("Top 10 Most popular Game Genres")
plt.xlabel("Number of Genres")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

## Free game vs Paid game

```python
# indicate which game is free
df['is_free'] = df['price'] == 0
df
```

```python
# create chart
plt.figure(figsize=(6,6))
sns.boxplot(data=df, x="is_free", y="average_2weeks")
plt.xticks([0,1], ['Paid', 'Free'])
plt.title("Free vs Paid Games: Player Count Distribution")
plt.ylabel("Playtime (Last 2 weeks)")
plt.xlabel("")
plt.tight_layout()
plt.show()
```
